```python
data = data.drop(['Left - right handed'], axis=1)
data = data.drop(['Music'], axis=1)
```

```python
corr_matrix = data.corr().abs()
corr_series = corr_matrix['Happiness in life']
for i in corr_series.index:
    if corr_series[i] < 0.10:
        corr_series.drop(labels = [i], inplace = True)
```

```python
x = list(corr_series.index)
x
```

```python
from sklearn.model_selection import train_test_split

x = data.drop(columns=x)
y = data['Happiness in life']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
print(x_train.shape) 
print(x_test.shape)
print(y_train.shape) 
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(x_train, y_train)
logreg.classes_
y_preds_logreg = logreg.predict(x_test)
y_preds_logreg
```

```python
from sklearn import metrics

def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

print('mean_absolute_error:')
print(metrics.mean_absolute_error(y_test,y_preds_logreg))
print('mean_squared_error:')
print(metrics.mean_squared_error(y_test,y_preds_logreg))
print('r2_score:')
print(metrics.r2_score(y_test,y_preds_logreg))
print('mean_absolute_percentage_error:')
print(mean_absolute_percentage_error(y_test,y_preds_logreg))
print('accuarcy score:')
print(metrics.accuracy_score(y_test,y_preds_logreg))
```
