# Simple Linear Regression

Train LinearRegression() model on your data.

```python
x = df.drop('diagnosis', axis=1)
y = df.diagnosis

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)

print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

l_r = LinearRegression()
l_r.fit(x_train, y_train)

y_pred_lr = np.round(l_r.predict(x_test))
y_pred_lr_list = []
for i in y_pred_lr:
    if i > 1:
        y_pred_lr_list.append(1)
    elif i < 0:
        y_pred_lr_list.append(0)
    else:
        y_pred_lr_list.append(int(i))
y_pred_lr = y_pred_lr_list

m = pd.Series(y_test, name='actual data').reset_index().drop('index', axis = 1)
n = pd.Series(y_pred_lr, name='predicted data').reset_index().drop('index', axis = 1)
visualization = pd.concat([m, n], axis = 1)
print(visualization['actual data'].value_counts())
print(visualization['predicted data'].value_counts())
print(visualization)
```

## Polynomial Regression

The polynomial features transform is available in the scikit-learn Python machine learning library via the PolynomialFeatures class. Apply the polynomial features transform, use different degrees.

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
p_f = PolynomialFeatures(3)
x_pf=p_f.fit_transform(x_train)

print(x_train.shape,x_pf.shape)
```

```python
l_r=LinearRegression()
l_r.fit(x_pf, y_train)
```

```python
p_f.transform(x_test).shape
```

```python
y_pred_lr_new = np.round(l_r.predict(p_f.transform(x_test)))
y_pred_lr_list_new = []
for i in y_pred_lr_new:
    if i > 1:
        y_pred_lr_list_new.append(1)
    elif i < 0:
        y_pred_lr_list_new.append(0)
    else:
        y_pred_lr_list_new.append(int(i))
y_pred_lr_new = y_pred_lr_list_new

m = pd.Series(y_test, name='actual data').reset_index().drop('index', axis = 1)
n = pd.Series(y_pred_lr_new, name='predicted data').reset_index().drop('index', axis = 1)
visualization = pd.concat([m, n], axis = 1)
print(visualization['actual data'].value_counts())
print(visualization['predicted data'].value_counts())
print(visualization)
```

# Metrics

Create a method print_metrics(), that prints the following metrics MAE (mean absolute error), MSE (mean squared error), MAPE (mean absolute percentage error), r2_score.

```python
from sklearn import metrics
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
print('Metrics for Linear Regression:')
print(mean_squared_error(y_test, y_pred_lr))
print(r2_score(y_test, y_pred_lr))
print(mean_absolute_error(y_test, y_pred_lr))
```

```python
print('Metrics for Polynomial Regression:')
print(mean_squared_error(y_test, y_pred_lr_new))
print(r2_score(y_test, y_pred_lr_new))
print(mean_absolute_error(y_test, y_pred_lr_new))
```
